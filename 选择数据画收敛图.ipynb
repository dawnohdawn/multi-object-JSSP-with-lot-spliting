```python
import pandas as pd
import os
import matplotlib.pyplot as plt
%matplotlib inline

PATH = os.path.abspath('.') + '\\modifiedGAModels\\convergeData'
```

##  originalMBO和originalMBO_newNeighbours的收敛对比

```python
def plotmbo(file1, file2):
    """
    功能：
    取data1的originalMBO，取data2的originalMBO_newNeighbours，画收敛曲线
    
    输入：
    file1，file2    要用到的csv文件全名
    """ 
    # 提取csv，取data1的originalMBO，取data2的originalMBO_newNeighbours
    data1 = pd.read_csv(PATH+"\\" + file1, header=0)
    data2 = pd.read_csv(PATH+"\\" + file2, header=0)
    # 混合需要的数据，在此可更改两个算法的名字
    mixdata = pd.DataFrame({'MBO':data1.originalMBO,'MBO+new neighbours':data2.originalMBO_newNeighbours})
    #画图
    mixdata.plot()
```

```python
# P1的理想图形
plotmbo('OMBO & OMBO+NEI-P1-87-90.csv', 'OMBO+NEI-P1-85.csv')
plt.savefig(PATH+"\\"+'MBO加入新的邻域算子后收敛对比-P1',dpi=160)

# 表现出收敛速度了
```

```python
# P2的理想图形
plotmbo('OMBO & OMBO+NEI-P2-188-184.csv', 'OMBO & OMBO+NEI-P2-188-184.csv')
plt.savefig(PATH+"\\"+'MBO加入新的邻域算子后收敛对比-P2',dpi=160)

# 表现出跳出局部最优了
```
