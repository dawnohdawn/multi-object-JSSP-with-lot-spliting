```python
import pandas as pd
import os
import matplotlib.pyplot as plt
%matplotlib inline

PATH = os.path.abspath('.') + '\\modifiedGAModels\\convergeData'
```

##  originalMBO和originalMBO_newNeighbours的收敛对比

```python
def plotmbo(file1, file2):
    """
    功能：
    取data1的originalMBO，取data2的originalMBO_newNeighbours，画收敛曲线
    
    输入：
    file1，file2    要用到的csv文件全名
    """ 
    # 提取csv，取data1的originalMBO，取data2的originalMBO_newNeighbours
    data1 = pd.read_csv(PATH+"\\" + file1, header=0)
    data2 = pd.read_csv(PATH+"\\" + file2, header=0)
    # 混合需要的数据，在此可更改两个算法的名字
    mixdata = pd.DataFrame({'MBO':data1.originalMBO,'MBO+nei':data2.originalMBO_newNeighbours})
    #画图
    mixdata.plot()
```

```python
# P1的理想图形
plotmbo('OMBO & OMBO+NEI-P1-87-90.csv', 'OMBO+NEI-P1-85.csv')
plt.savefig(PATH+"\\"+'MBO & MBO+nei 收敛对比-P1',dpi=160)

# 表现出收敛速度了
```

```python
# P2的理想图形
plotmbo('OMBO & OMBO+NEI-P2-188-184.csv', 'OMBO & OMBO+NEI-P2-188-184.csv')
plt.savefig(PATH+"\\"+'MBO & MBO+nei 收敛对比-P2',dpi=160)

# 表现出跳出局部最优了
```

## 多种群算法对比

```python
def plotmbo(file1, file2, file3, file4):
    """
    功能：
    取data1的originalMBO，取data2的originalMBO_newNeighbours，画收敛曲线
    
    输入：
    file1，file2    要用到的csv文件全名
    """ 
    # 提取csv，取data1的originalMBO，取data2的originalMBO_newNeighbours
    data1 = pd.read_csv(PATH+"\\" + file1, header=0)
    data2 = pd.read_csv(PATH+"\\" + file2, header=0)
    data3 = pd.read_csv(PATH+"\\" + file3, header=0)
    data4 = pd.read_csv(PATH+"\\" + file4, header=0)
    # 混合需要的数据，在此可更改两个算法的名字
    mixdata = pd.DataFrame({'ML-CCMBO':data1.myPMBO1,'CCMBO+stage':data2.myPMBO1,'CCMBO+diff':data3.myPMBO1,'PCCMBO':data4.myPMBO1})
    #画图
    mixdata.plot()
```

```python
# P1的理想图形
plotmbo('MLCCMBO-P1-83.csv','PCCMBO+stage-P1-85.csv', 'PCCMBO+stage-P1-85.csv', 'PCCMBO+stage-P1-85.csv')  # MBO & myMBO1-P1-85-85
plt.savefig(PATH+"\\"+'ML-CCMBO & CCMBO+stage & CCMBO+diff & PCCMBO收敛对比-P1',dpi=160)
```
